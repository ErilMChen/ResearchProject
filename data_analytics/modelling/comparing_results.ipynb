```python
import pandas as pd
```

```python
# Read random forest final metrics into dataframe
rfr_metrics = pd.read_csv("rfr_final_metrics.csv")
```

```python
rfr_metrics.sort_values(by="route", inplace=True)
rfr_metrics.head()
```

```python
rfr_metrics.shape
```

```python
# Read multi-layer perceptron metrics into dataframe
mlp_metrics = pd.read_csv("mlp_final_metrics.csv")
```

```python
mlp_metrics.sort_values(by="route", inplace=True)
mlp_metrics.head()
```

```python
mlp_metrics.shape
```

```python
merged_results = rfr_metrics.merge(mlp_metrics, on=["route"])
```

```python
merged_results.head()
```

```python
# These routes will not be included
merged_results.loc[(merged_results["MAE_actual_vs_planned_x"] < merged_results["5_fold_cv_MAE_x"]) & (merged_results["MAE_actual_vs_planned_y"] < merged_results["5_fold_cv_MAE_y"]) ]
```

```python
mlp_better = merged_results.loc[merged_results["5_fold_cv_MAE_y"] < merged_results["5_fold_cv_MAE_x"]]
```

```python
mlp_better.shape
```

```python
mlp_better.head()
```

```python
for index, row in mlp_better.iterrows():
    print(row["5_fold_cv_MAE_x"] - row["5_fold_cv_MAE_y"])
```

```python
routes_to_replace = mlp_better.route.unique().tolist()
routes_to_replace.sort()
```
