<p>Drop unusable features.</p>

```python
weather_all = df.drop(["dt_iso", "timezone", "city_name", "lat", "lon", "sea_level", "grnd_level", "rain_1h", "rain_3h", "snow_1h", "snow_3h", "weather_id", "weather_description", "weather_icon"], axis=1)
```

```python
weather_all.head()
```

```python
weather_all.dtypes
```

```python
def convert_to_timestamp(num):
    stamp = pd.Timestamp(num, unit="s")
    return stamp
```

```python
weather_all["dt"] = weather_all["dt"].apply(convert_to_timestamp)
```

```python
weather_all.head()
```

```python
df.dtypes
```

```python
weather_all.drop_duplicates(subset=["dt"], inplace=True)
```
